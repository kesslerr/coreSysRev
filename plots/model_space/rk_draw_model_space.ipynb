```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
sns.set_style('white')
```

### function to draw arrows

```python
def draw_connecting_arrow(ax, circ1, rad1, circ2, rad2, col = "k", lstyle = "-"):
    theta = np.arctan2(circ2[1] - circ1[1],
                       circ2[0] - circ1[0])

    starting_point = (circ1[0] + rad1 * np.cos(theta),
                      circ1[1] + rad1 * np.sin(theta))

    length = (circ2[0] - circ1[0] - (rad1 + 1.4 * rad2) * np.cos(theta),
              circ2[1] - circ1[1] - (rad1 + 1.4 * rad2) * np.sin(theta))

    ax.arrow(starting_point[0], starting_point[1], length[0], length[1],
             length_includes_head=True,
             head_width =.1,
             head_length=.15,
             color = col,
             linestyle = lstyle)
    
    # return the start- and end coordinates of the arrow for later usage
    return ()
             
```

```python

# node properties
radius = .2
Xofa   = .5
Yofa   = 1.
Xffa   = 1.5
Yffa   = .5
Xsts   = 1.5
Ysts   = 1.5

# text properties
fontSize = 5
font = {'family': 'sans', # was serif
        'color':  'black',
        'weight': 'normal', # 'normal'
        'size': fontSize,
        }
fontBold = {'family': 'sans', # was serif
        'color':  'black',
        'weight': 'bold', # 'normal'
        'size': fontSize,
        }

f, axarr = plt.subplots(4, 6, dpi=200)

modelCounter = 0
for i in range(4):
    for j in range(6):
        modelCounter += 1
        
        # OFA
        circle1=plt.Circle((Xofa,Yofa),radius,color=[55/255,126/255,184/255])
        axarr[i,j].add_patch(circle1)
        
        # FFA
        circle2=plt.Circle((Xffa,Yffa),radius,color=[77/255,175/255,74/255])
        axarr[i,j].add_patch(circle2)

        # STS
        circle3=plt.Circle((Xsts,Ysts),radius,color=[152/255,78/255,163/255])
        axarr[i,j].add_patch(circle3)

        # draw the node labels only in the first figure
        if (i==0) and (j==0):  
            axarr[i,j].text(Xofa-radius, Yofa+radius*1.2, 'OFA', fontdict=fontBold)
            axarr[i,j].text(Xffa+radius*1.2, Yffa-radius/2, 'FFA', fontdict=fontBold)
            axarr[i,j].text(Xsts+radius*1.2, Ysts-radius/4, 'STS', fontdict=fontBold)
        
        # draw the number of the model
        axarr[i,j].text(Xffa-Xofa, Ysts-Yffa, "#" + str(modelCounter), fontdict=font)
        
        axarr[i,j].set_xlim([0,2])      # set x lim
        axarr[i,j].set_ylim([0,2])      # set y lim
        axarr[i,j].set_aspect(1.)       # set equal aspect ratio
        axarr[i,j].axis('off')          # hide axes incl. tics and labels
        
        # connections of prototypes
        
        # OFA > FFA
        if (j == 0) or (j == 1) or (j == 3) or (j == 4) or (j == 5): 
            draw_connecting_arrow(axarr[i,j], (Xofa, Yofa), radius, (Xffa, Yffa), radius*0.8) # die 0.8 ist empirisch ausgewaehlt, komischerweise wuerde der Pfeil sonst nicht ganz den Node beruehren
        # FFA > STS
        if (j == 0) or (j == 4) or (j == 5):
            draw_connecting_arrow(axarr[i,j], (Xffa, Yffa), radius, (Xsts, Ysts), radius*0.8)        
        # OFA > STS
        if (j == 1) or (j == 2) or (j == 3) or (j == 4) or (j == 5): 
            draw_connecting_arrow(axarr[i,j], (Xofa, Yofa), radius, (Xsts, Ysts), radius*0.8)
        # STS > FFA
        if (j == 2) or (j == 3) or (j == 5):
            draw_connecting_arrow(axarr[i,j], (Xsts, Ysts), radius, (Xffa, Yffa), radius*0.8)              
        
        # C-matrix input arrows
        draw_connecting_arrow(axarr[i,j], (Xofa - 0.9, Yofa), radius, (Xofa, Yofa), radius*0.8)
        
        # additional backward connections
        if j == 0:
            if (i == 1) or (i == 3):
                draw_connecting_arrow(axarr[i,j], (Xffa, Yffa), radius, (Xofa, Yofa), radius*0.8)
            if (i == 2) or (i == 3):
                draw_connecting_arrow(axarr[i,j], (Xsts, Ysts), radius, (Xffa, Yffa), radius*0.8)

        if (j == 1) or (j == 3) or (j == 4) or (j == 5):
            if (i == 1) or (i == 3):
                draw_connecting_arrow(axarr[i,j], (Xffa, Yffa), radius, (Xofa, Yofa), radius*0.8)
            if (i == 2) or (i == 3):
                draw_connecting_arrow(axarr[i,j], (Xsts, Ysts), radius, (Xofa, Yofa), radius*0.8)
            
        if j == 2:
            if (i == 1) or (i == 3):
                draw_connecting_arrow(axarr[i,j], (Xffa, Yffa), radius, (Xsts, Ysts), radius*0.8)
            if (i == 2) or (i == 3):
                draw_connecting_arrow(axarr[i,j], (Xsts, Ysts), radius, (Xofa, Yofa), radius*0.8)
        
        # set titles for models
        # for the prototypes
        if i == 0:
            axarr[i,j].set_title('prototype ' + str(j+1), fontsize = fontSize)
        if j == 0:
            if i == 0:
                axarr[i,j].text(Xofa-radius*10, Yofa-radius*.5, 'no backward\nconnections', fontdict=font)
            if i == 1:
                axarr[i,j].text(Xofa-radius*10, Yofa-radius*.5, 'FFA backward\nconnection', fontdict=font)
            if i == 2:
                axarr[i,j].text(Xofa-radius*10, Yofa-radius*.5, 'STS backward\nconnection', fontdict=font)
            if i == 3:
                axarr[i,j].text(Xofa-radius*10, Yofa-radius*.5, 'both backward\nconnections', fontdict=font)
        
        # add box around model (for winning model and original winning model)
        if ( modelCounter == 2 ) or ( modelCounter == 24 ):
            axarr[i,j].plot([Xofa - radius - 0.1, Xofa - radius - 0.1],
                            [Yffa - radius - 0.1, Ysts + radius + 0.1], 'k--', lw=1.5) # left
            axarr[i,j].plot([Xofa - radius - 0.1, Xsts + radius + 0.1],
                            [Yffa - radius - 0.1, Yffa - radius - 0.1], 'k--', lw=1.5) # downer
            axarr[i,j].plot([Xofa - radius - 0.1, Xsts + radius + 0.1],
                            [Ysts + radius + 0.1, Ysts + radius + 0.1], 'k--', lw=1.5) # upper
            axarr[i,j].plot([Xsts + radius + 0.1, Xsts + radius + 0.1],
                            [Yffa - radius - 0.1, Ysts + radius + 0.1], 'k--', lw=1.5) #right
        # add label to winning models (original and new)
        if modelCounter == 2:
            axarr[i,j].text(Xofa - radius, Yffa-radius, 'original\nwinning model', fontdict=font)
        if modelCounter == 24:
            axarr[i,j].text(Xofa - radius, Yffa-radius, 'new\nwinning model', fontdict=font)

# adjust the spacing between subplots, so that there is not so much white space between subplots
plt.subplots_adjust(wspace=-0.05, hspace=-0.05)
   

f.savefig('/Users/Roman/projects/coreSysRev/plots/model_space/model_space_fairhallishai_dpi1000.png', bbox_inches='tight', dpi = 1000)
```

```python

fontSmall = {'family': 'sans', # was serif
        'color':  'black',
        'weight': 'normal', # 'bold'
        'size': 6,
        }
fontSmallBold = {'family': 'sans', # was serif
        'color':  'black',
        'weight': 'bold', # 'bold'
        'size': 8,
        }
radius = .15

def plot_model(A, Bf, Be, Asig, Bfsig, Besig, C, study, title = False, regions=False, label = ""):
    '''
    A  = 3x3 input ndarray
    Bf = 3x3 input ndarray
    Be = 3x3 input ndarray
    '''
    
    fig, ax = plt.subplots(1, 3, dpi=200)
    #greycolor = "0.1"
    for i in range(3):
        
        ax[i].set_xlim([0,2])      # set x lim
        ax[i].set_ylim([0,2])      # set y lim
        ax[i].set_aspect(1.)       # set equal aspect ratio
        ax[i].axis('off')          # hide axes incl. ticks and labels
        
        
        if Be == False:
            if i == 2:
                continue
            
        
        # OFA
        circle1=plt.Circle((Xofa,Yofa),radius,color=[55/255,126/255,184/255])
        ax[i].add_patch(circle1)

        # FFA
        circle2=plt.Circle((Xffa,Yffa),radius,color=[77/255,175/255,74/255])
        ax[i].add_patch(circle2)

        # STS
        circle3=plt.Circle((Xsts,Ysts),radius,color=[152/255,78/255,163/255])
        ax[i].add_patch(circle3)

        # input (C)
        


        
        
    if regions:
        ax[0].text(Xofa-radius-0.05, Yofa+radius*1.2, 'OFA', fontdict=fontSmallBold)
        ax[0].text(Xffa+radius*1.3, Yffa-radius/2, 'FFA', fontdict=fontSmallBold)
        ax[0].text(Xsts+radius*1.3, Ysts-radius/4, 'STS', fontdict=fontSmallBold)
        
    if title:
        ax[0].set_title('A-matrix\n(context-independent)', fontsize = 10) # title
        ax[1].set_title('C-matrix &\nB-matrix (faces)', fontsize = 10) # title
        ax[2].set_title('B-matrix\n(emotion)', fontsize = 10) # title

    # study label left    
    ax[0].text(-0.5 , .7, label, fontdict=fontSingle, rotation=90) # FFA -> OFA
    
    # input / C-matrix
    draw_connecting_arrow(ax[1], (Xofa - 0.9, Yofa), radius, (Xofa, Yofa), radius*0.8)
    ax[1].text(Xofa-.5 , Yofa+0.1 , C, fontdict=fontSmall) # FFA -> OFA
    
    # faces / A-matrix
    draw_connecting_arrow(ax[0], (Xofa, Yofa-0.15), radius, (Xffa, Yffa-0.15), radius*0.8, col = Asig[1,0]) # OFA > FFA
    draw_connecting_arrow(ax[0], (Xffa+0.1, Yffa),  radius, (Xsts+0.1, Ysts),  radius*0.8, col = Asig[2,1]) # FFA > STS        
    draw_connecting_arrow(ax[0], (Xofa, Yofa+.15),  radius, (Xsts, Ysts+.15),  radius*0.8, col = Asig[2,0]) # OFA > STS
    draw_connecting_arrow(ax[0], (Xsts, Ysts),      radius, (Xffa, Yffa),      radius*0.8, col = Asig[1,2]) # STS > FFA   
    draw_connecting_arrow(ax[0], (Xffa, Yffa),      radius, (Xofa, Yofa),      radius*0.8, col = Asig[0,1]) # FFA > OFA
    draw_connecting_arrow(ax[0], (Xsts, Ysts),      radius, (Xofa, Yofa),      radius*0.8, col = Asig[0,2]) # STS > OFA
    ax[0].text(Xffa - Xofa-.05 , Yofa - .2,A[0,1], fontdict=fontSmall) # FFA -> OFA
    ax[0].text(Xffa - Xofa -.3, Yofa - .5, A[1,0], fontdict=fontSmall) # OFA -> FFA
    ax[0].text(Xffa - Xofa -.3, Yofa + .4, A[2,0], fontdict=fontSmall) # OFA -> STS
    ax[0].text(Xffa - Xofa -.1, Yofa + .1, A[0,2], fontdict=fontSmall) # STS -> OFA
    ax[0].text(Xffa + 0.2 , Ysts-Yffa,     A[2,1], fontdict=fontSmall) # FFA -> STS
    ax[0].text(Xffa - 0.3 , Ysts-Yffa+0.0, A[1,2], fontdict=fontSmall) # STS -> FFA

    # faces / B-matrix
    draw_connecting_arrow(ax[1], (Xofa, Yofa-0.15), radius, (Xffa, Yffa-0.15), radius*0.8, col = Bfsig[1,0]) # OFA > FFA
    draw_connecting_arrow(ax[1], (Xffa+0.1, Yffa),  radius, (Xsts+0.1, Ysts),  radius*0.8, col = Bfsig[2,1]) # FFA > STS        
    draw_connecting_arrow(ax[1], (Xofa, Yofa+.15),  radius, (Xsts, Ysts+.15),  radius*0.8, col = Bfsig[2,0]) # OFA > STS
    draw_connecting_arrow(ax[1], (Xsts, Ysts),      radius, (Xffa, Yffa),      radius*0.8, col = Bfsig[1,2]) # STS > FFA   
    draw_connecting_arrow(ax[1], (Xffa, Yffa),      radius, (Xofa, Yofa),      radius*0.8, col = Bfsig[0,1]) # FFA > OFA
    draw_connecting_arrow(ax[1], (Xsts, Ysts),      radius, (Xofa, Yofa),      radius*0.8, col = Bfsig[0,2]) # STS > OFA
    ax[1].text(Xffa - Xofa-.05 , Yofa - .2,Bf[0,1], fontdict=fontSmall) # FFA -> OFA
    ax[1].text(Xffa - Xofa -.3, Yofa - .5, Bf[1,0], fontdict=fontSmall) # OFA -> FFA
    ax[1].text(Xffa - Xofa -.3, Yofa + .4, Bf[2,0], fontdict=fontSmall) # OFA -> STS
    ax[1].text(Xffa - Xofa -.1, Yofa + .1, Bf[0,2], fontdict=fontSmall) # STS -> OFA
    ax[1].text(Xffa + 0.2 , Ysts-Yffa,     Bf[2,1], fontdict=fontSmall) # FFA -> STS
    ax[1].text(Xffa - 0.3, Ysts-Yffa+0.0, Bf[1,2], fontdict=fontSmall) # STS -> FFA

    if Be != False:
        # emotion / B-matrix
        draw_connecting_arrow(ax[2], (Xofa, Yofa-0.15), radius, (Xffa, Yffa-0.15), radius*0.8, col = Besig[1,0]) # OFA > FFA
        draw_connecting_arrow(ax[2], (Xffa+0.1, Yffa),  radius, (Xsts+0.1, Ysts),  radius*0.8, col = Besig[2,1]) # FFA > STS        
        draw_connecting_arrow(ax[2], (Xofa, Yofa+.15),  radius, (Xsts, Ysts+.15),  radius*0.8, col = Besig[2,0]) # OFA > STS
        draw_connecting_arrow(ax[2], (Xsts, Ysts),      radius, (Xffa, Yffa),      radius*0.8, col = Besig[1,2]) # STS > FFA   
        draw_connecting_arrow(ax[2], (Xffa, Yffa),      radius, (Xofa, Yofa),      radius*0.8, col = Besig[0,1]) # FFA > OFA
        draw_connecting_arrow(ax[2], (Xsts, Ysts),      radius, (Xofa, Yofa),      radius*0.8, col = Besig[0,2]) # STS > OFA
        ax[2].text(Xffa - Xofa-.05 , Yofa - .2,Be[0,1], fontdict=fontSmall) # FFA -> OFA
        ax[2].text(Xffa - Xofa -.3, Yofa - .5, Be[1,0], fontdict=fontSmall) # OFA -> FFA
        ax[2].text(Xffa - Xofa -.3, Yofa + .4, Be[2,0], fontdict=fontSmall) # OFA -> STS
        ax[2].text(Xffa - Xofa -.1, Yofa + .1, Be[0,2], fontdict=fontSmall) # STS -> OFA
        ax[2].text(Xffa + 0.2 , Ysts-Yffa,     Be[2,1], fontdict=fontSmall) # FFA -> STS
        ax[2].text(Xffa - 0.3 , Ysts-Yffa+0.0, Be[1,2], fontdict=fontSmall) # STS -> FFA

    plt.show()
    fig.savefig('/Users/Roman/projects/coreSysRev/plots/BMA_per_study_'+study+'_dpi1000.png', bbox_inches='tight', dpi = 1000)
```
